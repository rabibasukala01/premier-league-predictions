```python

# removing 90s data
for files in os.listdir('PL_data'):
    if files.startswith('9'):
        os.remove(os.path.join('PL_data',files))
```

```python
def concat_df():
    # Directory containing the CSV files
    csv_directory = 'PL_data'

    # List all CSV files in the directory
    csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]

    # List to hold all DataFrames
    df_list = []

    # Read each CSV file and append it to the list
    for csv_file in csv_files:
        file_path = os.path.join(csv_directory, csv_file)
        try:
            print(f"Reading {csv_file} with UTF-8 encoding.")
            df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
            df_list.append(df)
        except UnicodeDecodeError:
            try:
                print(f"UTF-8 failed. Reading {csv_file} with ISO-8859-1 encoding.")
                df = pd.read_csv(file_path, encoding='ISO-8859-1', on_bad_lines='skip')
                df_list.append(df)
            except UnicodeDecodeError:
                print(f"Error decoding {csv_file}: Unable to decode file with UTF-8 or ISO-8859-1 encoding.")

    return df_list

# Concatenate all DataFrames into a single DataFrame
df_list = concat_df()
df_combined = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame to a new CSV file
df_combined.to_csv('combined_data.csv', index=False)

print('All CSV files have been appended into a single file.')
```
